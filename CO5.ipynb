```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/sample_data/Reviews.csv',engine='python',error_bad_lines=False)
```

```python
df.head(2)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Score'].value_counts()
```

```python
df['positively rated']=np.where(df['Score']>=3,1,0)
```

```python
df.head(2)
```

```python
df['positively rated'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df["Text"],df["positively rated"],random_state=50)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect=CountVectorizer(min_df=5,ngram_range=(1,2),).fit(X_train)
```

```python
vect.get_feature_names()
```

```python
len(vect.get_feature_names())
```

```python
X_train_vectorized=vect.transform(X_train)
```

```python
print(X_train_vectorized)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_vectorized,y_train)
```

```python
prediction=model.predict(vect.transform(X_test))
```

```python
print(prediction)
```

```python
print(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,prediction))
print(classification_report(y_test,prediction))
```

```python
from sklearn.metrics import roc_auc_score
print("AUC score is",roc_auc_score(y_test,prediction))
```
